```python
import pandas as pd
from datetime import datetime

def convert_to_datetime(date_string):
    try:
        # Convert to datetime object using strptime
        datetime_object = datetime.strptime(str(date_string), '%Y%m%d')

        # Format to 'dd-mmm-yy' using strftime
        formatted_date = datetime_object.strftime('%Y-%m-%d')

        return formatted_date
    except ValueError:
        # Handle cases where parsing fails
        return pd.NaT


```

```python
import openpyxl
import pandas as pd

# Load the Excel file
wb = openpyxl.load_workbook('/content/FR HVaR Report - Jan08, 2025_RP.xlsx')

# Access the 'Run Stats' sheet
sheet = wb['Run Stats']

# Load the report data into a pandas DataFrame
report = pd.DataFrame(sheet.values)

# Assuming the first row contains headers, set them
report.columns = report.iloc[0]
report = report[1:]

# ... (rest of your code for filtering and updating remains the same)

# Iterate over rows in filtered_noti DataFrame
for _, row in filtered_noti.iterrows():
    asset_class = row['Asset-Class']
    cob = row['COB']

    # Find matching rows in report DataFrame
    matching_rows = report[(report['Asset-Class'] == asset_class) & (report['COB'] == cob)]

    # Update values in the Excel sheet if a match is found
    if not matching_rows.empty:
        print(f"Found matching rows for Asset-Class: {asset_class}, COB: {cob}")
        for report_index in matching_rows.index:
            # Get the corresponding row number in the Excel sheet
            excel_row = report_index + 1  # +1 to account for header row and 1-based indexing in Excel

            # Iterate over columns in filtered_noti (excluding 'Asset-Class' and 'COB')
            for col_name in filtered_noti.columns:
                if col_name not in ['Asset-Class', 'COB']:
                    # Update the cell value if the column exists in the 'Run Stats' sheet
                    if col_name in report.columns:
                        sheet.cell(row=excel_row, column=report.columns.get_loc(col_name) + 1).value = row[col_name]

    else:
        print(f"No matching rows found for Asset-Class: {asset_class}, COB: {cob}")

    from openpyxl.styles import NamedStyle, Font, PatternFill, Border, Side, Alignment
    from openpyxl.styles.numbers import FORMAT_DATE_DATETIME, FORMAT_DATE_YYYYMMDD2

    if col_name == 'COB':
        sheet.cell(row=excel_row, column=report.columns.get_loc(col_name) + 1).number_format = 'yyyy-mm-dd'  # Or any desired format


# Save the changes to the Excel file
wb.save('/content/FR HVaR Report - Jan08, 2025_RP.xlsx')
```

```python
import pandas as pd
import numpy as np

# Assuming dq_target_df is your DataFrame

# Create the pivot table without margins initially
pivot_table = pd.pivot_table(
    dq_target_df,
    index=['level_4', 'dq_rule_applied', 'dq_rule_description'],  # Rows
    columns='is_full_reval',  # Columns
    values='count (uid)',  # Values
    aggfunc=lambda x: ' '.join(x.astype(str)),  # Custom aggregation to join as string
    fill_value='',  # Fill missing values with empty string
)

# For Grand Total to work
# Replace any non-numeric values with NaN in 'count (uid)' column for Grand Total calculation
dq_target_df['count (uid)'] = pd.to_numeric(dq_target_df['count (uid)'], errors='coerce')

# Calculate Grand Total separately for each column
grand_total_n = dq_target_df[dq_target_df['is_full_reval'] == 'n']['count (uid)'].sum()
grand_total_y = dq_target_df[dq_target_df['is_full_reval'] == 'y']['count (uid)'].sum()

# Add Grand Total row to pivot_table
pivot_table.loc[('Grand Total', '', ''), 'n'] = grand_total_n
pivot_table.loc[('Grand Total', '', ''), 'y'] = grand_total_y

# Reset index to bring index levels as columns
pivot_table = pivot_table.reset_index()

# --- Shift Grand Total row to the first row ---
# Get the index of the Grand Total row
grand_total_index = pivot_table[pivot_table['level_4'] == 'Grand Total'].index[0]

# Reorder the DataFrame using reindex
new_order = [grand_total_index] + [i for i in pivot_table.index if i != grand_total_index]
pivot_table = pivot_table.reindex(new_order)

# Reset index to consecutive numbers
pivot_table = pivot_table.reset_index(drop=True)

# Display the pivot table
print(pivot_table)
```
